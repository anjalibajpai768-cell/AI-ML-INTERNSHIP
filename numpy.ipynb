```python
arr = np.array([10,20,30,40,50])
print(np.mean(arr))   #to find the average
print(np.median(arr))  #to find median value
print(np.std(arr))   #standard deviation
print(np.max(arr))  #highest value
print (np.min(arr) ) #lowest value
print(np.sum(arr))  #total value
print(np.cumsum(arr)) #cumulative sum
print(np.var(arr))  #variance
# Corrected: To reshape an array, you need to provide an array and a shape compatible with its size.
# For example, to reshape 'arr' (5 elements) into a (5,1) column vector:
print(arr.reshape(5,1))
print(arr.flatten())
```
